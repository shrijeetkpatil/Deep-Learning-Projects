# Import Libraries

```python
import os
import shutil
from tqdm import tqdm

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.utils import shuffle 
from sklearn import metrics
from sklearn.metrics import confusion_matrix


%tensorflow_version 2.x
import tensorflow as tf
from tensorflow.keras import applications
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Flatten, Dense, Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import optimizers

# tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
sns.set_style("whitegrid") # For Graph Background White on Colab
```

# Define Image Data Generator

```python
train_datagen = ImageDataGenerator(rescale = 1./255, rotation_range=30, width_shift_range=0.3, height_shift_range=0.3, 
                                   brightness_range=None, shear_range=0.0, zoom_range=0.3, 
                                   channel_shift_range=0.0, fill_mode='nearest', horizontal_flip=True, 
                                   vertical_flip=False, validation_split=0.0, dtype='float32')
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255, rotation_range=30, width_shift_range=0.3, height_shift_range=0.3, 
                                  brightness_range=None, shear_range=0.0, zoom_range=0.3, 
                                  channel_shift_range=0.0, fill_mode='nearest', horizontal_flip=True, 
                                  vertical_flip=False, validation_split=0.0, dtype='float32')
```

# Generate Train and Test Image Data

```python
train_directory = './seg_train/seg_train/'
test_directory = './seg_test/seg_test/'

n_batch_size = 256
n_epochs = 50
```

```python
train_generator = train_datagen.flow_from_directory(train_directory, 
                                                   target_size=(150, 150), 
                                                   batch_size = n_batch_size, 
                                                   class_mode = 'categorical')
```

```python
test_generator = test_datagen.flow_from_directory(test_directory, 
                                                   target_size=(150, 150), 
                                                   batch_size = n_batch_size,
                                                   class_mode = 'categorical')
```

```python
train_labels = train_generator.classes
test_labels = test_generator.classes
class_names = list(test_generator.class_indices.keys())
```

# Display Some Train Images

```python
def display_train_data(class_names, images, labels, n_rows=5 , n_cols=5 , fig_width=10, fig_height=10):
    """
        Display n images from the images array with its corresponding labels
    """
    n_images = n_rows * n_cols

    fig = plt.figure(figsize=(fig_width, fig_height))
    fig.suptitle("First "+ str(n_images) +" Images of the Dataset", fontsize=16)
    for i in range(n_images):
        plt.subplot(n_rows,n_cols,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i], cmap=plt.cm.binary)
        plt.xlabel(class_names[np.where(labels[i] == 1)[0][0]])
    plt.show()
```

# Load VGG16 as Base Model

```python
vgg_model = applications.VGG16(weights='imagenet', include_top=False, input_shape=(150,150,3))

print('Model Loaded.')
```

# Create Custom Layers Top of on VGG16 Model

```python
vgg_model_output = vgg_model.output

top_model = Conv2D(filters=32, kernel_size=(3, 3), activation='relu', padding="same")(vgg_model_output)
top_model = Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding="same")(top_model)
top_model = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(top_model)
top_model = Dropout(0.25)(top_model)

top_model = Flatten()(top_model)

top_model = Dense(256, activation='relu')(top_model)
top_model = Dropout(0.5)(top_model)

top_model = Dense(512, activation='relu')(top_model)
top_model = Dropout(0.5)(top_model)

top_model = Dense(6, activation='softmax')(top_model)

custom_model = Model(inputs=vgg_model.input, outputs=top_model)
```

# Active Last 14 Layers

```python
for layer in custom_model.layers[:-14]:
    layer.trainable = False

for layer in custom_model.layers:
    print(layer, layer.trainable)
```

# Compile Our Model

```python
custom_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

# Generate Directory for Save Checkpoints after every Epoch

```python
# shutil.rmtree(checkpoint_dir)
```

```python
# Save The Checkpoints

# Include the epoch in the file name (uses `str.format`)
cp_dir = 'saved_checkpoints/'
if not os.path.exists(cp_dir):
    os.mkdir(cp_dir)

checkpoint_path = cp_dir+"cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
```

```python
# Create a callback that saves the model's weights every 5 epochs
cp_callback = ModelCheckpoint(
                  filepath=checkpoint_path, 
                  verbose=0, 
                  save_weights_only=True,
                  save_freq=5)
# Save the weights using the `checkpoint_path` format
custom_model.save_weights(checkpoint_path.format(epoch=0))
```

# Start Training

```python
trained_history = custom_model.fit( train_generator, steps_per_epoch = len(train_generator), 
                          epochs = n_epochs, use_multiprocessing=False, validation_data = test_generator, 
                          workers = 1, validation_steps=len(test_generator), callbacks = [cp_callback])
```

# Plot Graph to Compair Train & Test Accuracy and Loss

```python
def plot_graph(train_history, train_acc, test_acc, title, x_label, y_label):
    plt.plot(train_history.history[train_acc], label='Train')
    plt.plot(train_history.history[test_acc], label='Test')
    plt.title(title)
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.legend(loc='lower left')
    plt.grid()
    plt.show()
```

```python
title = 'Model Accuracy'
x_label = 'Epoch'
y_label = 'Accuracy'
plot_graph(trained_history, 'accuracy', 'val_accuracy', title, x_label, y_label)
```

```python
title = 'Model Loss'
x_label = 'Epoch'
y_label = 'Loss'
plot_graph(trained_history, 'loss', 'val_loss', title, x_label, y_label)
```

# Trained Model Apply on New (or Unseen) Data

## Load New Data

```python
IMAGE_SIZE = (150, 150)
```

```python
def load_new_data(datasets):
    
    images = []
   
    print("\nLoading {}".format(datasets))   
        # Iterate through each image in our folder
    for file in tqdm(os.listdir(datasets)):
                
                # Get the path name of the image
        img_path = os.path.join(datasets, file)
                
                # Open and resize the img
        image = cv2.imread(img_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, IMAGE_SIZE) 
                
        # Append the image and its corresponding label to the output
        images.append(image)
        
    images = np.array(images, dtype = 'float32')

    return images
    
```

```python
new_images = load_new_data('./seg_pred/seg_pred/')
```

```python
# Scale the Data
new_images = new_images / 255.0 
```

## Predict New Data

```python
predictions_new_data = custom_model.predict(new_images)
pred_labels_new_data = np.argmax(predictions_new_data, axis = 1)
```

## Display Some Predicted Images with Labels

```python
def display_data(class_names, images, labels, n_rows=5 , n_cols=5 , fig_width=10, fig_height=10):
    """
        Display n images from the images array with its corresponding labels
    """
    n_images = n_rows * n_cols

    fig = plt.figure(figsize=(fig_width, fig_height))
    fig.suptitle("First "+ str(n_images) +" Images of the Dataset", fontsize=16)
    for i in range(n_images):
        plt.subplot(n_rows,n_cols,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i], cmap=plt.cm.binary)
        plt.xlabel(class_names[labels[i]])
    plt.show()
```

```python
display_data(class_names, new_images, pred_labels_new_data, n_rows=7, n_cols=7 , fig_width=15, fig_height=15)
```

## Save the Predicted Images in their Class Folders

```python
def classify_new_images(class_names, images, labels):

    dirs = 'new_data_output'
    if not os.path.exists(dirs):
        os.mkdir(dirs)

    for i in range(len(images)):
    
        class_nm = class_names[labels[i]]

        class_path = os.path.join(dirs, class_nm)

        if not os.path.exists(class_path):
            os.mkdir(class_path)


        filename = class_path+'/'+class_nm+str(f"{i:04}")+'.jpg'
        cv2.imwrite(filename, images[i])
```

```python
classify_new_images(class_names, new_images*255.0 , pred_labels_new_data)
```
